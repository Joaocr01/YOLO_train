```python
from ultralytics import YOLO
```

```python
model = YOLO("yolov8m.pt")
```

```python
# Defina os argumentos de treinamento
training_args = {
    'data': 'data.yaml',  # Caminho para o arquivo de configuração de dados
    'epochs': 5           # Número de epochs
}

# Chame o método train com os argumentos expandidos
model.train(**training_args)
```
